## Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

# categorical data, hence using bar plot
plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.5)
plt.show()
```

```python
# using count plot for showing the ration in cancelations and bokings not cancelled

plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(handles=ax1.patches[:2], labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.show()
```

```python
# Find the percentage of bookings being cancelled in both the hotels.

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

#### Check if price is the reason of bookings being cancelled

```python
# grouping the reservations per day to take teh average value of it.

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (15, 8))
plt.title('Average Daily Rate in Resort Hotels', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
# Extract the month
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize = (20,10))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(handles=ax1.patches[:2], labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 20)
filtered_df = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index() # for taking out month from index, reset the index
sns.barplot(x='month', y='adr', data=filtered_df)
plt.xlabel('Month')
plt.ylabel('ADR (Average Daily Rate)')
plt.show()
```

```python
canceled_data = df[df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]

# value_counts is how many times country is appearing in the same column - returns the value in descending order
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

##### Now, compare the adr of the cancelled bookings and not cancelled bookings.

```python
cancelled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

##### It is again a proof that the Average Daily Rate is the factor that affects the cancellation and results in higher cancellation ratio.

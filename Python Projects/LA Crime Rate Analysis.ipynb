## Load data

```python
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv('Data for python course/Crime_Data_from_2020_to_Present.csv')
```

```python
# Check nulls
crime.isna().sum()
```

```python
df = crime.drop(['modus_operandi', 'weapon_code', 'weapon_description', 
                'crime_code_2', 'crime_code_3', 'crime_code_4', 'cross_street'], axis = 1)
df.head()
```

```python
df.sort_values('date_reported').head()
```

```python
# Update dtypes
df['date_reported'] = pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
```

```python
df['date_occurred'] = pd.to_datetime(df['date_occurred']).dt.strftime('%Y-%m-%d') # first coverted this column as string then converted to date time format
```

```python
df['date_occurred'] = pd.to_datetime(df['date_occurred'], format = '%Y-%m-%d')
```

```python
df.info()
```

```python
# Descriptive statistics
df.describe(include = ['float64', 'int64'])
```

```python
# victime age is falling in negatives, so let's check the counts of it. 

df['victim_age'].value_counts()
```

```python
# need to filter age greater than 0 
df_age = df[df['victim_age'] > 0]      #filter should be appplied inside []
```

```python
df_age['victim_age'].value_counts()
```

```python
df['Year'] = df['date_occurred'].dt.year
df['Month'] = df['date_occurred'].dt.month
```

```python
df.sort_values(by = 'date_reported')   # we can see by 4 dec, the data incoming stopped, hence the drop of crimes in december.
```

## Feature understanding and Analysis

```python
df['Month'].value_counts().sort_index().plot()
```

```python
df['Year'].value_counts().sort_index().plot(kind = 'bar')
```

```python
# Remove the crime date from December 2023 being incomplete.

df = df[~((df['Month'] == 12) & (df['Year'] == 2023))]
```

```python
df.sort_values(by = 'date_occurred')      # so,the data is now updated until 30 Nov, 2023
```

```python
# Using month and year cols, we can create a line chart that shows how many records are the in per month year combination.

df.groupby(['Year', 'Month']).size()
```

#### Find most common crimes that happens in LA

```python
crime_count_per_day = df.groupby(['date_occurred', 'crime_description']).size()
```

```python
# avergae of the crime count per day according to crime_description
top_20_crimes = crime_count_per_day.groupby('crime_description').mean().reset_index(name = 'average_daily_count')\
                        .sort_values(by = 'average_daily_count', ascending = False).head(20)
```
